```python
from scipy import optimize
from math import sin, exp
import matplotlib.pyplot as plt


def f(x):
    return sin(x / 5.0) * exp(x / 10.0) + 5 * exp(-x / 2.0)

# task 1
firstAnswer = optimize.minimize(f, 2, method='BFGS').fun
secondAnswer = optimize.minimize(f, 30, method='BFGS').fun
af = open('task1.txt', 'w')
af.write(" ".join(map(lambda x: str(round(x, 2)), [firstAnswer, secondAnswer])))
af.close()
% cat task1.txt
X = range(0, 40)
plt.plot(X, map(lambda x: f(x), X))
```

```python
# task 2
globalMin = optimize.differential_evolution(f, [(1, 30)]).fun
af = open('task2.txt', 'w')
af.write(str(round(globalMin, 2)))
af.close()
%cat task2.txt
```

```python
def h(x):
    return int(f(x))

X = range(0, 40)
plt.plot(X, map(lambda x: h(x), X))

firstAnswer = optimize.minimize(h, 30, method='BFGS').fun
secondAnswer = optimize.differential_evolution(h, [(1, 30)]).fun
af = open('task3.txt', 'w')
af.write(" ".join(map(lambda x: str(x), [firstAnswer, secondAnswer])))
af.close()
%cat task3.txt
```
